```python

# for INDEED WEBSITE
import requests
import bs4
from bs4 import BeautifulSoup
import pandas as pd
import time
max_results_per_city = 50
city_set = ['Hyderabad%2C+Telangana','hyderbad%2C+Telangana','Chennai%2C+Tamil+Nadu','Bangalore%2C+Karnataka',
            'Bengaluru%2C+Karnataka','Noida%2C+Uttar+Pradesh','Pune%2C+Maharashtra','Delhi','Delhi%2C+Delhi',
            'Gurgaon%2C+Haryana']
col = ['JOB_TITLE', 'COMPANY_NAME', 'LOCATION', 'SUMMARY']
comp=['Wipro','ICS Consultancy Services','Oracle','Accenture','IBM','Infosys','TCS','Amazon',
      'Tech Mahindra','Swiggy',' L&T infotech','Flipkart','Ola','Uber','Zomato','Cognizant']
sample_df = pd.DataFrame(columns = col)
res={}
res1={}
#scraping code:
for city in city_set:
    for start in range(0, max_results_per_city,10):
        o=0
        page = requests.get("https://www.indeed.co.in/jobs?q=software+developer&l=" + str(city) + "&start=" + str(start))
        time.sleep(1)  #ensuring at least 1 second between page grabs
        soup = BeautifulSoup(page.text, 'html.parser')
        for div in soup.find_all(name='div', attrs={'class':'row'}):
    #specifying row num for index of job posting in dataframe
            num = (len(sample_df) + 1) 
    #creating an empty list to hold the data for each posting
            job_post = [] 
    #grabbing job title
            for a in div.find_all(name='a', attrs={'data-tn-element':'jobTitle'}):
                         job_post.append(a['title']) 
    #grabbing company name
            company = div.find_all(name='span', attrs={'class':'company'}) 
            if len(company) > 0:
                        for b in company:
                            if b.text.strip() in comp:
                                job_post.append(b.text.strip())
                                if b.text.strip() in res.keys():
                                    res[b.text.strip()]+=1
                                else:
                                    res[b.text.strip()]=1
            else: 
                        sec_try = div.find_all(name='span', attrs={'class':'result-link-source'})
                        for span in sec_try:
                            if span.text.strip() in comp:
                                o=1
                                job_post.append(span.text.strip()) 
                                if span.text.strip() in res.keys():
                                    res[b.text.strip()]+=1
                                else:
                                    res[b.text.strip()]=1
    #grabbing location name
            c = div.findAll('span', attrs={'class': 'location'}) 
            for span in c:
                    job_post.append(span.text.strip())
                    if span.text.strip() in res1.keys() and o==1:
                            res1[span.text.strip()]+=1
                    else:
                            res1[span.text.strip()]=1
    #grabbing summary text
            d = div.findAll('div', attrs={'class': 'summary'}) #it conatins the exprience for the job
            for span in d:
                        s=span.text.strip()
                        s1=s.replace('\n',"")
                        job_post.append(s1)   
    #appending list of job post info to dataframe at index num
            if(len(job_post)>3):
                sample_df = sample_df.append({'JOB_TITLE' :job_post[0], 'COMPANY_NAME' :job_post[1], 'LOCATION' : job_post[2],
                                              'SUMMARY':job_post[3]},ignore_index = True)
print(sample_df)
print()
print("NO OF SOFTWARE DEVELOPERS JOBS BASED ON COMAPNIES")
for i in res:
    print(i,"    ",res[i],end="\n")
print()
print("NO OF SOFTWARE DEVELOPERS JOBS BASED ON CITIES")
for j in res1:
    print(j,"    ",res1[j],end="\n")
```
